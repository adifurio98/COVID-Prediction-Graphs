```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os
import time

# Import API key
gkey = {"Insert Google API Key here"}
```

```python
gmaps.configure(api_key=gkey)
```

```python
infections_df = pd.read_csv("Data/Cleaned/infections_cleaned.csv")
deaths_df = pd.read_csv("Data/Cleaned/deaths_cleaned2.csv")
cia_df = pd.read_csv("Data/Cleaned/cia_simplified.csv")
```

```python
cia_df["pop_density"] = round(cia_df["Population"] / cia_df["geography_area_total_sq_km"])
cia_df
```

```python
deaths_df2 = pd.read_csv("Data/Cleaned/deaths_totals.csv")
deaths_df3 = deaths_df2.dropna()
deaths_df3
```

```python
infect_totals3 = pd.read_csv("Data/Cleaned/infections_totals.csv")
infect_totals4 = infect_totals3.dropna()
infect_totals4
```

```python
country_stats = cia_df.drop(columns=["current_health_expenditure_2019", "physicians_density_per_1000_2020_to_2018", "Hospital_bed_density_per_1000_2018_to_2017", "real_GDP_per _capita_3year_average_2020_2018", "GDP_official_exchange_rate_Billion", "Real_GDP_purchasing_power_parity_3year_ave"])
country_stats
```

```python
covid_both = pd.merge(deaths_df3, infect_totals4, on="country")
covid_both
```

```python
all_totals = pd.merge(covid_both, country_stats, on="country")
all_totals['total_deaths'] = all_totals['total_deaths'].astype(float)
all_totals['total_infections'] = all_totals['total_infections'].astype(float)
pd.options.display.float_format = '{:,}'.format
all_totals
```

```python
country_list = list(all_totals["country"])
print(country_list)
```

```python
for country in country_list:
    target_url = f"https://maps.googleapis.com/maps/api/geocode/json?address={country}&key={gkey}"
    response = requests.get(target_url)
    print(response.status_code)

    geo_data = response.json()

    lat = geo_data["results"][0]["geometry"]["location"]["lat"]
    lng = geo_data["results"][0]["geometry"]["location"]["lng"]

    print(f"""Country: {country}
    Latitude: {lat}
    Longitude: {lng}""")
```

```python
rows = []

# assume null
lat = np.nan
lng = np.nan
name = None

for indx, row in all_totals.iterrows():
    new_row = {} # flat dictionary
    
    # Target city
    target_country = f"{row.country}"
    print(target_country)
    
    # Build the endpoint URL
    target_url = f"https://maps.googleapis.com/maps/api/geocode/json?address={target_country}&key={gkey}"
    
    try:

        # Run a request to endpoint and convert result to json
        response = requests.get(target_url)
        if(response.status_code == 200):

            geo_data = response.json()

            # Extract latitude and longitude
            if (len(geo_data["results"]) > 0):
                lat = geo_data["results"][0]["geometry"]["location"]["lat"]
                lng = geo_data["results"][0]["geometry"]["location"]["lng"]

                # geocoordinates
                target_coordinates = f"{lat}, {lng}"
                #target_search = "international airport"
                #target_radius = 500000
                #target_type = "airport"

                # set up a parameters dictionary
                params = {
                    "location": target_coordinates,
                    #"keyword": target_search,
                    #"radius": target_radius,
                    #"type": target_type,
                    "key": gkey
                }

                # base url
                base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
            
                # run a request using our params dictionary
                response = requests.get(base_url, params=params)
                if(response.status_code == 200):

                    data = response.json()

                    if (len(data["results"]) > 0):
                        # extract data
                        name = data["results"][0]["name"]
                        #vicinity = data["results"][0]["vicinity"]
                        #rating = data["results"][0]["rating"]
                        #user_ratings_total = data["results"][0]["user_ratings_total"]
            
    except Exception as e:
        print(e)
        
        # nuke it, everything null
        lat = np.nan
        lng = np.nan
        name = None
        #vicinity = None
        #rating = np.nan
        #user_ratings_total = np.nan
                    
    # flatten data and append
    new_row["country"] = target_country
    new_row["latitude"] = lat
    new_row["longitude"] = lng
    #new_row["airport"] = name
    #new_row["address"] = vicinity
    #new_row["rating"] = rating
    #new_row["user_ratings_total"] = user_ratings_total
    
    rows.append(new_row)
    
    # let api breathe
    time.sleep(2)
```

```python
df = pd.DataFrame(rows)
df
```

```python
master = pd.merge(all_totals, df, on="country")
master
```

```python
infos = []

for indx, row in master.iterrows():
    info_box_template = f"""<dl>
                        <dt>Name:</dt><dd>{row.country}</dd>
                        <dt>Total Deaths:</dt><dd>{row.total_deaths}</dd>
                        <dt>Total Infections:</dt><dd>{row.total_infections}</dd>
                        <dt>Population Density (People per Sq Km):</dt><dd>{row.pop_density}</dd>
                        </dl>
                        """
    infos.append(info_box_template)
    
infos[0]
```

```python
# get data
data = master.loc[:, ["latitude", "longitude"]]

# Create BASE, init the graph
figure_layout = {
    'width': '800px',
    'height': '600px',
    'border': '0px',
    'padding': '1px',
    'margin': '0 auto 0 auto'
}
fig = gmaps.figure(layout=figure_layout)

# init the overlay layers
markers = gmaps.marker_layer(data, info_box_content=infos)
fig.add_layer(markers)

# show the plot
fig
```
