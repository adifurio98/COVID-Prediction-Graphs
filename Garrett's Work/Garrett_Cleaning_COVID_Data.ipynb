```python
# This Notebook was created to clean the JHU COVID data for deaths and infections, and create CSV files for future use. 

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
# Read CSVs for COVID Data

infections_df = pd.read_csv("Data/JHU_COVID/CONVENIENT_global_confirmed_cases.csv")
deaths_df = pd.read_csv("Data/JHU_COVID/CONVENIENT_global_deaths.csv")
```

```python
deaths_df.head()
```

```python
# Cols to keep: "Country/Region", "Austria", "France" (has 11), "Portugal", "Greece", "Netherlands" (has 4), "Indonesia", "India", "Mexico", "Thailand", "Iran"

infections_df2 = infections_df.drop(infections_df.index[0])
infections_df2.reset_index(drop=True, inplace=True)
infections_df2.head()
```

```python
infections_df3 = infections_df2.loc[:, ["Country/Region", "Austria", "Italy", "Portugal", "Greece", "Taiwan*", "Indonesia", "India", "Mexico", "Thailand", "Iran"]]
infections_df3.head()
```

```python
infections_df3.info()
```

```python
infections_df3.rename(columns={'Country/Region': 'Date'}, inplace=True)
infections_df3.head()
```

```python
infections_df3.to_csv('infections_cleaned.csv', index=False)
```

```python
deaths_df2 = deaths_df.drop(deaths_df.index[0])
deaths_df2.reset_index(drop=True, inplace=True)
deaths_df2.head()
```

```python
deaths_df3 = deaths_df2.loc[:, ["Country/Region", "Austria", "Italy", "Portugal", "Greece", "Taiwan*", "Indonesia", "India", "Mexico", "Thailand", "Iran"]]
deaths_df3.head()
```
