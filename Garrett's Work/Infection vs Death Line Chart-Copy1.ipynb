```python
# This Notebook is a copy of Alyssa's Notebook, with some edits made to actually display our initial ten countries of interest,
# and show line charts for those countries daily deaths and infections over the course of the pandemic.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
infection = "Data/Cleaned/infections_cleaned.csv"
death = "Data/Cleaned/deaths_cleaned2.csv"
infectiond = pd.read_csv(infection)
deathd = pd.read_csv(death)
infectiondf = pd.DataFrame(infectiond)
deathdf = pd.DataFrame(deathd)

infectiondf["Date"] = pd.to_datetime(infectiondf["Date"])
deathdf["Date"] = pd.to_datetime(deathdf["Date"])
```

```python
infectiondf['Austria'].mask(infectiondf['Austria'] < 0, 0, inplace=True)
infectiondf['Italy'].mask(infectiondf['Italy'] < 0, 0, inplace=True)
infectiondf['Portugal'].mask(infectiondf['Portugal'] < 0, 0, inplace=True)
infectiondf['Greece'].mask(infectiondf['Greece'] < 0, 0, inplace=True)
infectiondf['Indonesia'].mask(infectiondf['Indonesia'] < 0, 0, inplace=True)
infectiondf['Taiwan*'].mask(infectiondf['Taiwan*'] < 0, 0, inplace=True)
infectiondf['India'].mask(infectiondf['India'] < 0, 0, inplace=True)
infectiondf['Mexico'].mask(infectiondf['Mexico'] < 0, 0, inplace=True)
infectiondf['Thailand'].mask(infectiondf['Thailand'] < 0, 0, inplace=True)
infectiondf['Iran'].mask(infectiondf['Iran'] < 0, 0, inplace=True)
```

```python
infectiondf["Day"] = infectiondf.Date.dt.day
```

```python
sub = infectiondf.loc[infectiondf.Day.isin([1,10, 20])]
sub.head()
```

```python
plt.figure(figsize=(10,6))
x = sub["Date"]
y = sub['Austria']
y2 = sub["Italy"]
y3 = sub["Portugal"]
y4 = sub["Greece"]
y5 = sub["India"]
y6 = sub["Taiwan*"]
y7 = sub["Indonesia"]
y8 = sub["Mexico"]
y9 = sub["Thailand"]
y10 = sub["Iran"]

plt.plot(x, y, label="Austria")
plt.plot(x, y2, label="Italy")
plt.plot(x, y3, label="Portugal")
plt.plot(x, y4, label="Greece")
plt.plot(x, y5, label="India")
plt.plot(x, y6, label="Taiwan")
plt.plot(x, y7, label="Indonesia")
plt.plot(x, y8, label="Mexico")
plt.plot(x, y9, label="Thailand")
plt.plot(x, y10, label="Iran")
plt.ylabel("Number of Infections")
plt.title("Covid Infection Cases")
plt.legend(loc='upper right')
plt.show()
plt.savefig("infection.png")
```
